```python
# Exercise #1
# Reverse the list below in-place using an in-place algorithm.
# For extra credit: Reverse the strings at the same time.

# Original words = ['this', 'is', 'a', 'sentence', '.']

def reverse(original, n):
    for i in range(0, int(n / 2)):
        original[i], original[n - i - 1] = original[n - i - 1], original[i]

if __name__=="__main__":
    original = ['this', 'is', 'a', 'sentence', '.']
    n =len(original)
    print(*original)
    reverse(original, n)
    print(*original)
```

```python
# # Exercise #2
# # Create a function that counts how many distinct words are in the string below, then outputs a dictionary with the words as the key and the value as the amount of times that word appears in the string.
# # Should output:
# # {'a': 5,
# # 'abstract': 1,
# # 'an': 3,
# # 'array': 2, ... etc..

a_text = 'In computing, a hash table hash map is a data structure which implements an associative array abstract data type, a structure that can map keys to values. A hash table uses a hash function to compute an index into an array of buckets or slots from which the desired value can be found'

def string(a_string):
    dict = {}
    list = a_text.split(" ")
    for words in list:
        temp = {words: list.count(words)}
        dict.update(temp)
    return dict

print(string(a_text))
```

```python
# Exercise #3
# Write a program to implement a Linear Search Algorithm. Also in a comment, write the Time Complexity of the following algorithm.

# Hint: Linear Searching will require searching a list for a given number.

# nums_list = [10, 23, 45, 70, 11, 15]
# target = 70

def linear(num_list, n, target):
    for i in range(0,n):
        if (num_list[i]==target):
            return i
    return -1

num_list = [10, 23, 45, 70, 11, 15]
target = 70
n = len(num_list)

result = linear(num_list, n, target)
if (result == -1):
    print(num_list.index(target))
else:
    print(result)
    
```
